## learning-AI101 : topic4_1

**RNN (Recurrent Neural Network)**  
- 보통은 Dense 레이어에 집어넣을 땐 input data는 한꺼번에 집어넣어지게 됨. 데이터의 순서 정보가 사라진다
- 따라서 순차적인 데이터를 딥러닝시킬 땐 방식이 달라야함
    - 순차 데이터 (시계열, Sequence) : 음성인식, 악보, 문장 ...
- 해결책 : simple RNN 레이어를 이용하여 input data를 순차적으로, 한번에 하나씩 넣을 수 있게끔 도와주자

<hr>

- RNN은 BPTT 방식을 이용해서 학습 효율을 이끌어나가는데,
    - BPTT : Back Propagation Through Time
    - 다음 input이 들어올 때 그 전 예측 결과의 일부를 사용한다 (hidden state)

RNN을 적용할 수 있는 예시  
- vector를 입력해서 일종의 Sequence를 예측할 때
    - 그림을 입력할 때, 이 그림의 상황을 문장으로 출력하는 인공지능
- Sequence를 입력해서 일종의 Vector를 예측할 때
    - 긴 문장들을 입력하고, 그 문장들을 한 그림으로 요약하는 인공지능
    - 네이버 클린봇
- Sequence를 입력해서 일종의 Sequence를 예측할 때
    - 네이버 파파고

<hr>

Simple RNN은 **diminishing gradient(경사 추락)** 문제로 잘 사용되지 않는다
- 만약 I love _____에서 밑줄을 예측하는 RNN을 구현한다고 했을 때
    - 첫번째 레이어에서 I를 이용해 예측한 결과를 a라고 하자
    - 두번째 레이어 예측 과정에서 a의 일부를 사용한다. 따라서 예측 결과를 b라고 하자
    - 세번째 레이어 예측 과정에서 b를 사용한다. 그럼 b에 들어있는 a의 정보는 엄청 적을 것. 인공지능이 a의 정보를 까먹는 것이다.

**따라서 문제를 해결하기 위해 LSTM을 사용한다**

- 다음 레이어로 예측 포인트가 이동될 때 hidden state 뿐만이 아닌 cell state (장기기억)을 가지고 간다
- 따라서 초반에 학습하였던 데이터의 기억까지 가지고 예측을 진행하기 때문에 diminishing gradient가 발생하지 않는다
