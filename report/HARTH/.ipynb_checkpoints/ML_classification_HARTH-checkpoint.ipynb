다음과 같이 모든 파일들의 records들을 병합한다.

- 1) ./data/harth를 path로 지정한다.
- 2) glob 라이브러리를 사용하여 특정 디렉토리에 있는 모든 CSV 파일의 경로를 가져온다.
- 3) 각 파일을 순회하면서 pandas의 read_csv 함수를 사용하여 파일을 읽고 데이터프레임에 추가한다.
- 4) 모든 데이터프레임을 하나로 병합한다.

````python
# records 병합

import glob

directory_path = './data/harth'

# 모든 CSV 파일의 경로를 가져오기
all_files = glob.glob(directory_path + "/*.csv")

print (all_files)

# 데이터프레임을 담을 빈 리스트 생성
df_list = []

# 모든 파일을 순회하며 데이터프레임에 추가
for filename in all_files :
    df = pd.read_csv(filename)  
    df_list.append(df) 

merged_df = pd.concat(df_list, ignore_index=True) # 병합 (header는 ignore)
df = merged_df

df.head() 
````

````python
# df의 크기 및 info
print("df 크기 : ", df.shape)
df.info()
````

#### 1-2) 전처리 (열 삭제, 결측치 제거, Z-score normalization)

````python
# timestamp, index 삭제, unnamed 삭제

try : 
    del df['timestamp']
    del df['index']
except : 
    pass

df.drop(['Unnamed: 0'], axis = 1, inplace = True)
print ("현재 attribute : ", df.columns)
````

records의 결측치를 찾기 위하여 아래와 같은 코드를 실행한다.

````python
# 결측치 확인

for feature in df.columns :
    print (f'{feature}열의 결측치 존재 여부 : ', df[feature].isnull())
    print ('\n--------------\n')

print ("데이터의 총 결측치 수 : ", df.isnull().sum())
````

위를 보면, 본 데이터에 결측치가 **전혀** 존재하지 않고 있다. 따라서 결측치에 따른 전처리를 진행하지 않아도 된다.

<br>

logistic regression을 통해 classification을 진행할 것이기에 StandardScaler을 통한 스케일링이 필요하다. StandardScaler를 통한 Z-score normalization을 수행한다.

````python
# input과 target으로 split

harth_input = df[['back_x', 'back_y', 'back_z', 'thigh_x', 'thigh_y', 'thigh_z']].to_numpy() 
harth_target = df['label'].to_numpy()
````

````python
# X_train, y_train, X_test, y_test으로 train set, test set 각각 8:2로 split

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(harth_input, harth_target, 
                                                   stratify=harth_target, 
                                                   test_size=0.2)
````

````python
# StandardScaler을 통한 정규화 진행

from sklearn.preprocessing import StandardScaler
ss_transformer = StandardScaler()

ss_transformer.fit(X_train)

X_train_scaled = ss_transformer.transform(X_train)
X_test_scaled = ss_transformer.transform(X_test)
````

### 3. logistic regression 모델을 적용하여 classification
- logistic regression 모델을 사용하여 다중 분류를 진행한다.
- softmax 함수를 사용하여 feature 수만큼의 z값을 확률로 변환한다.
    - 직접 z값을 구하여 진행할 수 있으나, predict_proba 메소드를 사용하여 진행하도록 한다.
- hyperparameter인 max_iter 최적값을 도출한다.
    - 최적의 hyperparameter를 찾기 위해 랜덤 서치 기법을 사용한다.
    - 그리드는 아래와 같이 정의한다
        - C : 1부터 50까지 randint
        - max_iter : 10부터 50까지 randint

````python
# logistic regression 모델 fit 및 train score, test score 출력
# 이때, max_iter 최적의 하이퍼파라미터를 찾기 위해 random search 기법 사용

from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from scipy.stats import randint

# 하이퍼파라미터 그리드 정의
param_grid = {
    'C': randint(1, 50),
    'max_iter': randint(10, 50)
}

# Logistic Regression 모델 정의
lr = LogisticRegression()

# Ramdom Search 설정
random_search = RandomizedSearchCV(lr, param_grid, n_iter=10, n_jobs=-1)

# fit
random_search.fit(X_train_scaled, y_train)

# 최적의 하이퍼파라미터 출력 및 score 출력
print(f'최적의 hyperparameter : {random_search.best_params_}')
print(f'cross-validation score : {random_search.best_score_}')
````

score가 높지 않은 것을 보아, underfitting된 것으로 보인다. 따라서 feature engineering을 통한 feature addition으로 전체적인 학습율을 높일 것이다.

````python
# PolynormialFeatures으로 특성 늘리기 -> 다시 스케일링

from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures()
poly.fit(X_train)

X_train_poly = poly.transform (X_train_scaled)
X_test_poly = poly.transform (X_test_scaled)

print (f"shape | X_train_poly : {X_train_poly.shape}, X_test_poly : {X_test_poly.shape}")
print (f"poly | {poly.get_feature_names_out()}")

ss_transformer = StandardScaler()

ss_transformer.fit(X_train_poly)

X_train_poly_scaled = ss_transformer.transform(X_train_poly)
X_test_poly_scaled = ss_transformer.transform(X_test_poly)
````

다시 grid search를 진행한다.

````python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from scipy.stats import randint

lr = LogisticRegression()
random_search = RandomizedSearchCV(lr, param_grid, n_iter=10, n_jobs=-1) # 특성이 더 늘어났기 때문에 n_iter를 2로 제한한다
random_search.fit(X_train_poly_scaled, y_train)

# 최적의 하이퍼파라미터 출력 및 score 출력
print(f'최적의 hyperparameter : {random_search.best_params_}')
print(f'cross-validation score : {random_search.best_score_}')
````

cross-validation score가 전에 비해서 확실히 늘은 것을 알 수 있다. 이를 시각화한 figure은 아래와 같다.

````python
# 하이퍼파라미터 조합별 성능 시각화

# 각 하이퍼파라미터 조합별 교차 검증 스코어 가져오기
results = random_search.cv_results_
mean_scores = results['mean_test_score']
params = results['params']

# 시각화를 위한 데이터 준비
C_values = [params[i]['C'] for i in range(len(params))]
max_iter_values = [params[i]['max_iter'] for i in range(len(params))]

# 가로 방향으로 막대 그래프로 교차 검증 스코어 시각화
fig, ax = plt.subplots(figsize=(8, 5))
bars = ax.barh(range(len(mean_scores)), mean_scores, tick_label=[f"C={C}, max_iter={max_iter}" for C, max_iter in zip(C_values, max_iter_values)])
ax.set_ylabel('Hyperparameters')
ax.set_xlabel('Mean CV Accuracy')
ax.set_title('Random Search Scores')
plt.xlim([0.7, 1.0])  # x 축 범위 설정 (필요에 따라 조정 가능)
plt.tight_layout()
plt.show()
````

따라서 'C': 21, 'max_iter': 36으로 설정하여 logistic regression 할 시 가장 score가 크다는 것을 알 수 있다. 아래에 학습과 score, predict가 이어진다.

````python
# fit, score

lr = LogisticRegression(C=21, max_iter=36)
lr.fit (X_train_poly_scaled, y_train)

lr_score = [lr.score(X_train_poly_scaled, y_train), lr.score(X_test_poly_scaled, y_test)]
print (f"train score, test score : {lr_score}")
````

````python
# 모델의 예측

predict_X = X_train_poly_scaled[0:101]  # feature data 중 상단 100개
prediction = lr.predict(predict_X)

print("prediction : ", prediction)

cnt = 0

# 예측 결과와 실제 라벨 비교
for i in range (100) : 
    print(f"predicted label: {prediction[i]}")
    print(f"actual label: {y_train[i]}")
    
    if prediction[i] == y_train[i]:
        cnt += 1
        
    print('--------------------')

print ('cnt : ', cnt)
print ('accuracy : ', cnt / 100)
````
