결국엔 맨 마지막 column인 Outcome 열이 target이 될 것이며, 나머지는 feature가 되어 classification을 진행해야 할 것이다. Outcome 열은 0 혹은 1로 이루어진 클래스 변수 (이진 변수)이며, 추후 편향 가능성 여부를 파악하기 위하여 countplot을 그려 확인해보자.

```python
# (4) heatmap

correlation_matrix = df[df.columns].corr() 
plt.figure(figsize=(10, 8)) 
sns.heatmap(correlation_matrix, annot=True, cmap='RdBu', square=True,
            linewidths=1, linecolor='black', cbar_kws={"shrink": 0.8})

plt.title('Correlation Heatmap of Diabetes Dataset', fontsize=16)
plt.show()
```

```python
# (5) feature, target data split, target countplot

diabetes_input = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']].to_numpy() 
diabetes_target = df['Outcome'].to_numpy()

print ("0, 1의 갯수 : ", len(df[df['Outcome'] == 0]), len(df[df['Outcome'] == 1]))

# countplot 시각화

sns.countplot(data=df, y=diabetes_target)
plt.show()
```

위 figure를 확인해보니, target 중 0이 1보다 약 2배 많다는 것을 확인 가능하다. **이를 통하여 편향이 발생하여 confusion matrix의 특정 diagonal component (0, 0)에 위치한 숫자가 높을 것 같다**고 생각한다. 추후에 **stritify 속성을 도입**하여 dataset을 split할 예정이다.

```python
# (6) train set, test set, validation set split, scaling (z-score normalization)

from sklearn.model_selection import train_test_split

# X_train, y_train, X_test, y_test
X_train, X_test, y_train, y_test = train_test_split(diabetes_input, 
                                                   diabetes_target,
                                                   test_size=0.2,
                                                   stratify=diabetes_target)

# X_train, y_train, X_val, y_val
X_train, X_val, y_train, y_val = train_test_split(X_train, 
                                                   y_train,
                                                   test_size=0.2,
                                                   stratify=y_train)

print (X_train.shape, 
      X_test.shape,
       X_val.shape)
```

```python
# (8) scaling

from sklearn.preprocessing import StandardScaler

# normal data scaling (Z-score normalization)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

<s>원래는 이 이후에, y의 값을 encoding하여 이진 변수로 나타내야 한다. 하지만 이미 Outcome 열은 0과 1만 존재하기 때문에, 기존 데이터를 그대로 사용하도 one-hot encoding된 상태에서 사용하는 것과 다름이 없다는 것이다. 따라서 one-hot encoding 과정을 생략한다.</s> 

**라고 생각하면 안된다.** one-hot encoding은 단순히 0과 1로만 나타내는 것이 아닌, ndarray형태로 [0, 1], [1, 0]과 같이 target을 변환하는 것이니 말이다. 따라서 one-hot encoding을 실시한다.

```python
# (10) df to torch

def to_tensor(data, dtype=torch.float32) -> torch.Tensor:
    """
    Convert pandas DataFrame or NumPy array to a PyTorch tensor.
    
    Parameters:
        data (pd.DataFrame, np.ndarray, or torch.Tensor): Input data
        dtype (torch.dtype): Desired tensor data type (default: torch.float32)
    
    Returns:
        torch.Tensor: Converted tensor
    """
    if isinstance(data, torch.Tensor):
        tensor = data.to(dtype)
    elif isinstance(data, pd.DataFrame):
        tensor = torch.tensor(data.values, dtype=dtype)
    elif isinstance(data, np.ndarray):
        tensor = torch.tensor(data, dtype=dtype)
    else:
        raise TypeError(f"Unsupported data type: {type(data)}")

    print("OK :", tensor.shape)
    return tensor

X_train_tensor = to_tensor(X_train_scaled)
y_train_tensor = to_tensor(y_train).reshape(-1, 1)

X_val_tensor = to_tensor(X_val_scaled)
y_val_tensor = to_tensor(y_val).reshape(-1, 1)

X_test_tensor = to_tensor(X_test_scaled)
y_test_tensor = to_tensor(y_test).reshape(-1, 1)
```

-------

### 4. DNN 
- 아래와 같이 experiment할 예정이다.
    - optimizer는 **적응형 학습율 optimizer인 adamW을 사용한다.**
    - validation loss 감소를 위하여 **dropout 층을 추가한다.**
        - callback으로 early stopping을 도입하지 않고, 충분히 학습하여 loss의 추이를 확인한다.

```python
# modelling DNN

from torchinfo import summary

class MLP (nn.Module) :
    def __init__(self) : 
        super().__init__()
        # Linear → BatchNorm → Activation → Dropout
        self.linear = nn.Sequential(
            nn.Linear(8, 50),
            nn.BatchNorm1d(50),
            nn.ReLU(),
            nn.Dropout(p=0.5),
            
            nn.Linear(50, 40),
            nn.BatchNorm1d(40),
            nn.ReLU(),
            nn.Dropout(p=0.4),
            
            nn.Linear(40, 30),
            nn.BatchNorm1d(30),
            nn.ReLU(),
            nn.Dropout(p=0.3),
            
            nn.Linear(30, 20),
            nn.BatchNorm1d(20),
            nn.ReLU(),
            nn.Dropout(p=0.2),
            
            nn.Linear(20, 10),
            nn.BatchNorm1d(10),
            nn.ReLU(),
            nn.Dropout(p=0.1),
            
            nn.Linear(10, 1),

            nn.Sigmoid()
        )
        
    def forward (self, x) : 
        x = self.linear(x)
        return x
    
model = MLP()
summary(model, 
       input_size=(1, 8))
```

```python
# compile

criterion = nn.BCELoss()
optimizer = optim.AdamW(model.parameters(), 
                        lr = 0.001)
epoch = 100
accuracy_history = []
loss_history = []
```

```python
# fit

for ep in range (epoch) : 
    model.train()
    # inference (prediction)
    y_hat = model.forward(X_train_tensor)
    
    # loss
    loss = criterion (y_hat, y_train_tensor)
    
    # deriviate
    optimizer.zero_grad()
    loss.backward()
    
    # update
    optimizer.step()
    
    train_loss = loss.item()
    train_acc, val_acc, val_loss = 0, 0, 0
    
    model.eval()
    with torch.no_grad() : 
        train_pred = model.forward(X_train_tensor)
        val_pred = model.forward(X_val_tensor)
        
        val_loss = criterion (val_pred, y_val_tensor)
        
        train_pred_class = (train_pred > 0.5).float()
        val_pred_class = (val_pred > 0.5).float()

        train_acc = (train_pred_class == y_train_tensor).float().mean()
        val_acc = (val_pred_class == y_val_tensor).float().mean()
    print (f"epoch {ep + 1} | train loss {train_loss:.4f}, val loss {val_loss:.4f}, train acc {train_acc:.2f}, val acc {val_acc:.2f}")
    
    accuracy_history.append ((train_acc, val_acc))
    loss_history.append ((train_loss, val_loss))
```

```python
# loss 시각화

plt.plot([ep+1 for ep in range (epoch)], 
         [tup[0] for tup in loss_history], 
        label="train_loss")
plt.plot([ep+1 for ep in range (epoch)], 
         [tup[1] for tup in loss_history], 
        label="val_loss")
plt.legend()
plt.show()
```

```python
# accuracy 시각화

plt.plot([ep+1 for ep in range (epoch)], 
         [tup[0] for tup in accuracy_history], 
        label="train_acc")
plt.plot([ep+1 for ep in range (epoch)], 
         [tup[1] for tup in accuracy_history], 
        label="val_acc")
plt.legend()
plt.show()
```

```python
# classification report and confusion matrix (시각화)


from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

model.eval()
with torch.no_grad():
    y_pred = model(X_test_tensor)
    y_pred_class = (y_pred > 0.5).int().view(-1).cpu().numpy()
    y_true_class = y_test_tensor.view(-1).cpu().numpy()


print(classification_report(y_true_class, y_pred_class))

cm = confusion_matrix(y_true_class, y_pred_class)

plt.figure(figsize=(5, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Pred 0', 'Pred 1'],
            yticklabels=['True 0', 'True 1'])
plt.xlabel("Prediction")
plt.ylabel("Ground Truth")
plt.title("Confusion Matrix")
plt.tight_layout()
plt.show()
```
