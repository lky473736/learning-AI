범주형 자료들을 출력해보았다. 이런 범주형 변수들까지 학습시키기 위해선, 추후 **LabelEncoder를 이용하여 각 고유한 값들을 전부 0과 1의 고유벡터로 전환시켜야 할 것**이다.
