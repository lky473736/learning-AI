```python
# def create_windows_from_data(df, window_size=500, stride=250):
#     X_windows = []
#     y_windows = []
#     subject_ids = []

#     signals_col = [feature for feature in df.columns if feature.startswith('signal_')]

#     for idx, row in df.iterrows():
#         subject_id = row['subject_id']
#         sbp = row['sBP']
#         dbp = row['dBP']
#         signal_data = row[signals_col].values.astype(np.float64)

#         for start in range(0, len(signal_data) - window_size + 1, stride):
#             window = signal_data[start:start + window_size]
#             if len(window) == window_size:
#                 X_windows.append(window)
#                 y_windows.append([sbp, dbp])
#                 subject_ids.append(subject_id)

#     return np.array(X_windows), np.array(y_windows), subject_ids

def create_windows_from_data(df, window_size=500, stride=250):
    X_windows = []
    y_windows = []
    subject_ids = []

    signals_col = [feature for feature in df.columns if feature.startswith('signal_')]

    for idx, row in df.iterrows():
        subject_id = row['subject_id']
        sbp = row['sBP']
        dbp = row['dBP']

        signal_data = row[signals_col].values.astype(np.float32)

        if np.isnan(signal_data).any():
            print(f"NaN found in subject {subject_id}, skipping")
            continue

        for start in range(0, len(signal_data) - window_size + 1, stride):
            window = signal_data[start:start + window_size]
            if len(window) == window_size:
                X_windows.append(window)
                y_windows.append([sbp, dbp])
                subject_ids.append(subject_id)

    return np.array(X_windows), np.array(y_windows), subject_ids

window_size = 100
stride = 50

X_train_windows, y_train_windows, train_subject_ids = create_windows_from_data(train_df, window_size, stride)
X_val_windows, y_val_windows, val_subject_ids = create_windows_from_data(val_df, window_size, stride)
X_test_windows, y_test_windows, test_subject_ids = create_windows_from_data(test_df, window_size, stride)

print(f"Training windows: {X_train_windows.shape}")
print(f"Validation windows: {X_val_windows.shape}")
print(f"Test windows: {X_test_windows.shape}")
```

```python
import torch.backends.cudnn as cudnn
cudnn.benchmark = True
cudnn.deterministic = False

class SSM(nn.Module):
    def __init__(self, d_model, d_state, seq_len):
        super().__init__()
        self.d_model = d_model
        self.d_state = d_state
        self.seq_len = seq_len

        self.in_proj = nn.Linear(d_model, d_model)

        init_scale = 0.0001  # 0.001    
        self.A = nn.Parameter(torch.randn(d_model, d_state, d_state) * init_scale)
        self.B = nn.Parameter(torch.randn(d_model, d_state, 1) * init_scale)
        self.C = nn.Parameter(torch.randn(d_model, 1, d_state) * init_scale)
        self.D = nn.Parameter(torch.randn(d_model) * init_scale)
        self.out_proj = nn.Linear(d_model, d_model)

    def forward(self, x):
      batch_size, seq_len, d_model = x.shape

      x_proj = self.in_proj(x)

      h = torch.zeros(batch_size, d_model, self.d_state, device=x.device)
      outputs = torch.zeros(batch_size, seq_len, d_model, device=x.device)

      for t in range(seq_len):
          u = x_proj[:, t, :].unsqueeze(-1).unsqueeze(-1)

          #    
          h_temp = torch.matmul(self.A, h.unsqueeze(-1)).squeeze(-1)
          h_temp = torch.clamp(h_temp, min=-1e6, max=1e6)  #   

          u_temp = torch.matmul(self.B, u).squeeze(-1)
          u_temp = torch.clamp(u_temp, min=-1e6, max=1e6)

          h = h_temp + u_temp

          y_temp = torch.matmul(self.C, h.unsqueeze(-1)).squeeze(-1).squeeze(-1)
          y_temp = torch.clamp(y_temp, min=-1e6, max=1e6)

          outputs[:, t, :] = y_temp + self.D * x_proj[:, t, :]

      outputs = self.out_proj(outputs)
      return outputs
```

```python
class MAMBABlock(nn.Module):
    def __init__(self, d_model, d_state, seq_len, expand_factor=2, dropout=0.1):
        super().__init__()
        self.d_model = d_model
        self.d_expanded = d_model * expand_factor

        self.norm1 = nn.LayerNorm(d_model)
        self.norm2 = nn.LayerNorm(d_model)

        self.expand_proj = nn.Linear(d_model, self.d_expanded)
        self.gate_proj = nn.Linear(d_model, self.d_expanded)

        self.ssm = SSM(self.d_expanded, d_state, seq_len)

        self.contract_proj = nn.Linear(self.d_expanded, d_model)

        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        residual = x

        x = self.norm1(x)

        expanded = self.expand_proj(x)
        gate = F.silu(self.gate_proj(x))
        gated_expanded = expanded * gate

        ssm_out = self.ssm(gated_expanded)

        out = self.contract_proj(ssm_out)

        out = self.dropout(out)
        out = out + residual

        out = self.norm2(out)

        return out
```

```python
class MAMBA_CNN_DNN_Windowed(nn.Module):
    def __init__(self, input_size, output_size, d_model=64, d_state=16,
                 num_mamba_blocks=2, cnn_channels=[64, 32], dnn_hidden=[128, 64, 32],
                 dropout=0.2):
        super().__init__()
        self.d_model = d_model
        self.seq_len = input_size

        self.input_proj = nn.Linear(1, d_model)

        self.mamba_blocks = nn.ModuleList([
            MAMBABlock(d_model, d_state, input_size, dropout=dropout)
            for _ in range(num_mamba_blocks)
        ])

        cnn_layers = []
        in_channels = d_model

        for out_channels in cnn_channels:
            cnn_layers.append(nn.Conv1d(in_channels, out_channels, kernel_size=3, padding=1))
            cnn_layers.append(nn.BatchNorm1d(out_channels))
            cnn_layers.append(nn.ReLU())
            cnn_layers.append(nn.MaxPool1d(kernel_size=2, stride=2))
            cnn_layers.append(nn.Dropout(dropout))
            in_channels = out_channels

        self.cnn = nn.Sequential(*cnn_layers)

        cnn_output_size = input_size // (2 ** len(cnn_channels))
        dnn_input_size = cnn_channels[-1] * cnn_output_size

        dnn_layers = []
        input_dim = dnn_input_size

        for hidden_dim in dnn_hidden:
            dnn_layers.append(nn.Linear(input_dim, hidden_dim))
            dnn_layers.append(nn.ReLU())
            dnn_layers.append(nn.Dropout(dropout))
            input_dim = hidden_dim

        dnn_layers.append(nn.Linear(input_dim, output_size))

        self.dnn = nn.Sequential(*dnn_layers)

    def forward(self, x):
        batch_size = x.shape[0]

        if x.dim() > 2:
            x = x.reshape(batch_size, self.seq_len)

        x = x.unsqueeze(-1)
        x = self.input_proj(x)

        for block in self.mamba_blocks:
            x = block(x)

        x = x.transpose(1, 2)
        x = self.cnn(x)
        x = x.flatten(1)
        x = self.dnn(x)

        return x
```

```python
input_size = window_size
output_size = 2
d_model = 64
d_state = 16
num_mamba_blocks = 2
cnn_channels = [64, 32]
dnn_hidden = [128, 64, 32]

model = MAMBA_CNN_DNN_Windowed(
    input_size=input_size,
    output_size=output_size,
    d_model=d_model,
    d_state=d_state,
    num_mamba_blocks=num_mamba_blocks,
    cnn_channels=cnn_channels,
    dnn_hidden=dnn_hidden,
    dropout=0.2
)
```

```python
def prepare_data(X_train, y_train, X_val, y_val, X_test, y_test, batch_size=8):
    X_train_tensor = torch.tensor(X_train.values, dtype=torch.float32)
    y_train_tensor = torch.tensor(y_train[['sBP', 'dBP']].astype(float).values, dtype=torch.float32)

    X_val_tensor = torch.tensor(X_val.values, dtype=torch.float32)
    y_val_tensor = torch.tensor(y_val[['sBP', 'dBP']].astype(float).values, dtype=torch.float32)

    X_test_tensor = torch.tensor(X_test.values, dtype=torch.float32)
    y_test_tensor = torch.tensor(y_test[['sBP', 'dBP']].astype(float).values, dtype=torch.float32)

    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)
    test_loader = DataLoader(test_dataset, batch_size=batch_size)

    return train_loader, val_loader, test_loader
```

```python
from torch.amp import autocast, GradScaler  #  import 
import gc

def train_model(model, train_loader, val_loader,
                epochs=50,
                lr=0.005,
                device='cuda'):
    model = model.to(device)
    criterion = nn.L1Loss()
    optimizer = optim.Adam(model.parameters(), lr=lr, weight_decay=1e-5)

    #   GradScaler 
    scaler = GradScaler('cuda')
    accumulation_steps = 4

    #  OneCycleLR   ( )
    scheduler = optim.lr_scheduler.OneCycleLR(
        optimizer,
        max_lr=lr,
        steps_per_epoch=len(train_loader) // accumulation_steps + 1,
        epochs=epochs,
        pct_start=0.3,
        anneal_strategy='cos'
    )

    # ReduceLROnPlateau     ( )
    # scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    #     optimizer,
    #     mode='min',
    #     factor=0.5,
    #     patience=3,
    #     min_lr=1e-7
    # )

    train_losses = []
    val_losses = []
    best_val_loss = float('inf')

    for epoch in range(epochs):
        model.train()
        train_loss = 0.0
        optimizer.zero_grad(set_to_none=True)

        for i, (inputs, targets) in enumerate(tqdm(train_loader, desc=f'Epoch {epoch+1}/{epochs}')):
            inputs, targets = inputs.to(device), targets.to(device)

            try:
                #   autocast 
                with autocast('cuda'):
                    outputs = model(inputs)
                    loss = criterion(outputs, targets) / accumulation_steps

                    # NaN 
                    if torch.isnan(loss).any():
                        print(f"NaN loss detected at batch {i}, skipping")
                        continue

                scaler.scale(loss).backward()

                #  NaN 
                valid_grads = True
                for name, param in model.named_parameters():
                    if param.grad is not None and torch.isnan(param.grad).any():
                        print(f"NaN gradient in {name}, skipping update")
                        valid_grads = False
                        break

                if not valid_grads:
                    optimizer.zero_grad(set_to_none=True)
                    continue

                if (i + 1) % accumulation_steps == 0 or (i + 1) == len(train_loader):
                    scaler.unscale_(optimizer)

                    #   
                    torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=0.1)

                    scaler.step(optimizer)
                    scaler.update()
                    scheduler.step()  # OneCycleLR  
                    optimizer.zero_grad(set_to_none=True)

                train_loss += (loss.item() * accumulation_steps) * inputs.size(0)

            except RuntimeError as e:
                print(f"Error in batch {i}: {e}")
                optimizer.zero_grad(set_to_none=True)
                continue

            del outputs, loss

        train_loss /= len(train_loader.dataset)
        train_losses.append(train_loss)

        gc.collect()
        torch.cuda.empty_cache()

        model.eval()
        val_loss = 0.0

        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs, targets = inputs.to(device), targets.to(device)

                with autocast('cuda'):
                    outputs = model(inputs)
                    loss = criterion(outputs, targets)

                val_loss += loss.item() * inputs.size(0)

                del outputs, loss

        val_loss /= len(val_loader.dataset)
        val_losses.append(val_loss)

        # ReduceLROnPlateau  ,  
        # scheduler.step(val_loss)

        print(f'Epoch {epoch+1}/{epochs}, Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}')

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), 'best_mamba_tcn_dnn_model.pt')
            print(f"model save (val loss: {val_loss:.4f})")

        gc.collect()
        torch.cuda.empty_cache()

    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.legend()
    plt.show()

    return model, train_losses, val_losses
```

```python
def evaluate_model(model, test_loader, device='cuda'):
    model = model.to(device)
    model.eval()
    criterion = nn.L1Loss()

    test_loss = 0.0
    predictions = []
    actuals = []

    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            with autocast():
                outputs = model(inputs)
                loss = criterion(outputs, targets)

            test_loss += loss.item() * inputs.size(0)

            predictions.append(outputs.cpu().numpy())
            actuals.append(targets.cpu().numpy())

            del outputs, loss

    test_loss /= len(test_loader.dataset)

    predictions = np.vstack(predictions)
    actuals = np.vstack(actuals)

    mae = np.mean(np.abs(predictions - actuals), axis=0)
    rmse = np.sqrt(np.mean((predictions - actuals)**2, axis=0))

    print(f'Test Loss: {test_loss:.4f}')
    print(f'MAE (sBP, dBP): {mae[0]:.2f}, {mae[1]:.2f}')
    print(f'RMSE (sBP, dBP): {rmse[0]:.2f}, {rmse[1]:.2f}')

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.scatter(actuals[:, 0], predictions[:, 0], alpha=0.6)
    plt.plot([actuals[:, 0].min(), actuals[:, 0].max()], [actuals[:, 0].min(), actuals[:, 0].max()], 'r--')
    plt.xlabel('Actual sBP')
    plt.ylabel('Predicted sBP')
    plt.title(f'Systolic BP (MAE: {mae[0]:.2f}, RMSE: {rmse[0]:.2f})')

    plt.subplot(1, 2, 2)
    plt.scatter(actuals[:, 1], predictions[:, 1], alpha=0.6)
    plt.plot([actuals[:, 1].min(), actuals[:, 1].max()], [actuals[:, 1].min(), actuals[:, 1].max()], 'r--')
    plt.xlabel('Actual dBP')
    plt.ylabel('Predicted dBP')
    plt.title(f'Diastolic BP (MAE: {mae[1]:.2f}, RMSE: {rmse[1]:.2f})')

    plt.tight_layout()
    plt.savefig('prediction_results.png')
    plt.close()

    return test_loss, mae, rmse, predictions, actuals
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print (f"device : {device}")

def prepare_windowed_data(X_windows, y_windows, batch_size=32):
    X_tensor = torch.tensor(X_windows, dtype=torch.float32)
    y_tensor = torch.tensor(y_windows, dtype=torch.float32)

    dataset = TensorDataset(X_tensor, y_tensor)
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

    return loader

batch_size = 32
train_loader = prepare_windowed_data(X_train_windows, y_train_windows, batch_size)
val_loader = prepare_windowed_data(X_val_windows, y_val_windows, batch_size)
test_loader = prepare_windowed_data(X_test_windows, y_test_windows, batch_size)


trained_model, train_losses, val_losses = train_model(model,
                                                      train_loader, val_loader,
                                                  epochs=50, lr=0.0001, device=device)

model.load_state_dict(torch.load('best_mamba_tcn_dnn_model.pt'))

test_loss, mae, rmse, predictions, actuals = evaluate_model(model, test_loader, device=device)

print(model)
```

```python
def evaluate_model(model, test_loader, device='cuda'):
    model = model.to(device)
    model.eval()
    criterion = nn.L1Loss()

    test_loss = 0.0
    predictions = []
    actuals = []

    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs, targets = inputs.to(device), targets.to(device)

            with autocast('cuda'):  #  
                outputs = model(inputs)
                loss = criterion(outputs, targets)

            test_loss += loss.item() * inputs.size(0)

            predictions.append(outputs.cpu().numpy())
            actuals.append(targets.cpu().numpy())

            del outputs, loss

    test_loss /= len(test_loader.dataset)

    predictions = np.vstack(predictions)
    actuals = np.vstack(actuals)

    mae = np.mean(np.abs(predictions - actuals), axis=0)
    rmse = np.sqrt(np.mean((predictions - actuals)**2, axis=0))

    return test_loss, mae, rmse, predictions, actuals
```

```python
model.load_state_dict(torch.load('best_mamba_tcn_dnn_model.pt'))
test_loss, mae, rmse, predictions, actuals = evaluate_model(model, test_loader, device=device)

print("=====    =====")
print(f'MAE (sBP): {mae[0]:.2f} mmHg')
print(f'MAE (dBP): {mae[1]:.2f} mmHg')
print(f'RMSE (sBP): {rmse[0]:.2f} mmHg')
print(f'RMSE (dBP): {rmse[1]:.2f} mmHg')
```

```python
def visualize_first_100_samples(predictions, actuals):
    samples = min(100, len(predictions))

    plt.figure(figsize=(15, 8))

    #  x
    indices = range(samples)

    #  
    plt.subplot(2, 1, 1)  # 2 1   
    # SBP   
    plt.plot(indices, actuals[:samples, 0], 'b-', linewidth=1.5, label='True SBP')
    plt.plot(indices, predictions[:samples, 0], 'orange', linewidth=1.5, label='Predicted SBP')
    plt.ylabel('SBP (mmHg)')
    plt.title('Systolic Blood Pressure - First 100 Samples')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.subplot(2, 1, 2)  # 2 1   
    # DBP   
    plt.plot(indices, actuals[:samples, 1], 'g-', linewidth=1.5, label='True DBP')
    plt.plot(indices, predictions[:samples, 1], 'r-', linewidth=1.5, label='Predicted DBP')
    plt.xlabel('Sample Index')
    plt.ylabel('DBP (mmHg)')
    plt.title('Diastolic Blood Pressure - First 100 Samples')
    plt.legend()
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('first_100_samples_prediction.png', dpi=300)
    plt.show()

#  100  
visualize_first_100_samples(predictions, actuals)
```
