오차함수 형태의 노이즈 데이터를 생성하고, GAN(Generative Adversarial Network)을 이용하여 이와 유사한 데이터를 생성해보겠다. GAN이란 Generator와 Discriminator가 서로 다른 목적을 가지고 경쟁하여 자기 스스로를 update하는 비지도 학습의 일종으로, 기존과 유사한 데이터를 생성함과 동시에 다양한 variation을 만든다.

- Generator : 생성자는 무작위 노이즈 벡터를 받아서, 별 모양의 노이즈를 만들도록 학습된다. 생성자는 점차적으로 점의 위치, 크기, 각도 등을 제어하여 훈련 데이터를 닮은 별 모양을 생성할 수 있게 된다.

- Discriminator : 판별자는 생성된 데이터를 실제 별 모양 데이터와 비교하면서 진짜와 가짜를 구별하는 역할을 한다. 훈련 데이터를 사용하여 판별자는 "진짜" 데이터와 "가짜" 데이터를 정확하게 구별할 수 있도록 학습한다.

위 함수 generate를 가지고 일정한 noise level의 오차함수 형태의 noise를 만들었다. 위 noise가 GAN에서 Discriminator에 대입될 실제 데이터이다. GAN에서 generator는 처음에 무작위 noise를 가지고 형편 없는 데이터를 생성해내고, Discriminator가 계속 피드백하면서 데이터 수정 정책을 고쳐나간다.

```python
def generate_real_samples(n):
    x1 = np.random.rand(n) - 0.5 
    x2 = (x1 - 2) * (x1 - 1) * x1 * (x1 + 1) * (x1 + 2) 
    
    x1 = x1.reshape(n, 1)  # x1을 (n, 1) 형태로 변환
    x2 = x2.reshape(n, 1)  # x2를 (n, 1) 형태로 변환

    X = np.hstack((x1, x2))  # x1과 x2를 결합하여 (x1, x2) 형태의 데이터를 만듬
    y = np.ones((n, 1))      # 실제 데이터는 y값 1로 설정 (당연한거임)

    return X, y
```

```python
def define_discriminator(inputs=2):
    model = Sequential(name='Discriminator')
    model.add(Dense(25, activation='relu', kernel_initializer='he_uniform', input_dim=inputs))  
    model.add(LeakyReLU(alpha=0.01))
    model.add(Dense(15, activation='relu', kernel_initializer='he_uniform'))  
    model.add(LeakyReLU(alpha=0.01))
    model.add(Dense(5, activation='relu', kernel_initializer='he_uniform'))  
    model.add(LeakyReLU(alpha=0.01))
    model.add(Dense(1, activation='sigmoid'))  # real(1) or fake(0)
    
    model.compile(optimizer=tf.optimizers.Adam(), loss='binary_crossentropy', metrics=['accuracy'])  
    return model
```

```python
discriminator_model = define_discriminator()
discriminator_model.summary()
```

```python
def define_generator(latent_dim, outputs=2):
    model = Sequential(name='Generator')
    model.add(Dense(25, activation='relu', kernel_initializer='he_uniform', input_dim=latent_dim))
    model.add(LeakyReLU(alpha=0.01))
    model.add(Dense(15, activation='relu', kernel_initializer='he_uniform'))
    model.add(LeakyReLU(alpha=0.01))
    model.add(Dense(outputs, activation='linear'))

    return model
```

```python
latent_dim = 5
generator_model = define_generator(latent_dim)
generator_model.summary()
```

```python
def generate_latent_points(latent_dim, n):
    '''generate points in latent space as input for the generator'''
    x_input = np.random.rand(latent_dim*n) #generate points in latent space
    x_input = x_input.reshape(n,latent_dim)  #reshape
    return x_input

def generate_fake_samples(generator, latent_dim, n):
    x_input = generate_latent_points(latent_dim, n) #genarate points in latent space
    x = generator.predict(x_input) #predict outputs
    y = np.zeros((n, 1))
    return x, y
```

위에 보이는 두 함수는 결국에 generator가 가짜 데이터를 생성하기 위한 준비를 하는 함수다.

`generate_latent_points` 함수는 생성자가 학습할 수 있도록 잠재 공간에서 랜덤한 입력 벡터를 생성하는 역할을 한다. 이 함수는 주어진 차원(latent_dim)과 개수(n)만큼의 랜덤 값을 생성한 후, 이를 생성자의 입력으로 사용할 수 있도록 재구성한다.

`generate_fake_samples` 함수는 생성자가 생성할 가짜 데이터를 만든다. 이 함수는 먼저 `generate_latent_points` 함수를 호출하여 잠재 공간에서 샘플을 생성하고, 그 샘플을 생성자에 입력하여 가짜 데이터를 예측한다. 그 후, 이 가짜 데이터를 "가짜"로 레이블링하기 위해 `y`를 0으로 설정하고, 생성된 가짜 데이터와 레이블을 반환한다.

```python
X, _ = generate_fake_samples(generator_model, latent_dim, 100)
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
def define_gan(generator, discriminator):
    '''define the combined generator and discriminator model'''
    discriminator.trainable = False
    model = Sequential()
    model.add(generator)
    model.add(discriminator)
    model.compile(optimizer = 'adam', loss = 'binary_crossentropy')
    return model
```

```python
gan_model = define_gan(generator_model, discriminator_model)
gan_model.summary()
```

```python
def summarize_performance(epoch, generator, discriminator, latent_dim, n = 100):
    '''evaluate the discriminator and plot real and fake samples'''
    x_real, y_real = generate_real_samples(n)      #generate real samples
    _, acc_real = discriminator.evaluate(x_real, y_real, verbose = 1)
    x_fake, y_fake = generate_fake_samples(generator, latent_dim, n)
    _, acc_fake = discriminator.evaluate(x_fake, y_fake, verbose = 1)
    print('Epoch: ' + str(epoch) + ' Real Acc.: ' + str(acc_real) + ' Fake Acc.: '+ str(acc_fake))
    plt.scatter(x_real[:,0], x_real[:,1], color = 'red')
    plt.scatter(x_fake[:,0], x_fake[:,1], color = 'blue')
    plt.show()
```

```python
def train_gan(g_model,d_model,gan_model,latent_dim, num_epochs = 10000,num_eval = 2000, batch_size = 128):
    ''' function to train gan model'''
    half_batch = int(batch_size/2)
  #run epochs
    for i in range(num_epochs):
        X_real, y_real = generate_real_samples(half_batch) #generate real examples
        d_model.train_on_batch(X_real, y_real)               # train on real data
        X_fake, y_fake = generate_fake_samples(g_model, latent_dim, half_batch) #generate fake samples
        d_model.train_on_batch(X_fake, y_fake)                #train on fake data
        #prepare points in latent space as input for the generator
        x_gan = generate_latent_points(latent_dim, batch_size)
        y_gan = np.ones((batch_size, 1))    #generate fake labels for gan
        gan_model.train_on_batch(x_gan, y_gan)
        if (i+1) % num_eval == 0:
            summarize_performance(i + 1, g_model, d_model, latent_dim)
```

```python
train_gan(generator_model, discriminator_model, gan_model, latent_dim)
```
