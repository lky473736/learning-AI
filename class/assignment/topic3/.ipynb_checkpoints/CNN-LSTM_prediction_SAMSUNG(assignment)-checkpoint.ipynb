````python
# df_train과 df_test로 분리, 'Date' 열 제거

# split to train data and test data
# train : test = 9 : 1

split_point = int(0.9 * df.shape[0])
print ("split point : ", split_point)

df_train = df[:split_point]
train_dates = dates[:split_point]

df_test = df[split_point:]
test_dates = dates[split_point:]

df_train = df_train.drop('Date', axis=1)
df_test = df_test.drop('Date', axis=1)

print (df_train)
print ()
print (df_test)
````

````python
# 결측치 제거

print("Train set missing values:\n", df_train.isnull().sum())
print("Test set missing values:\n", df_test.isnull().sum())

df_train = df_train.dropna()
df_test = df_test.dropna()

print("After dropping missing values in train set:\n", df_train.isnull().sum())
print("After dropping missing values in test set:\n", df_test.isnull().sum())
````

````python
# normalize the dataset

scaler = StandardScaler()
scaler = scaler.fit(df_train)
df_train_scaled = scaler.transform(df_train)
df_test_scaled = scaler.transform(df_test)

print (df_train_scaled)
print (df_test_scaled)
````

위와 같은 형식으로 데이터를 구성한다. 위 block에는 window size를 k라고 하였을 때, k개의 records가 input, 그 다음 Close가 target이 되는 형태로 구성되었다. 사실 이것이 순환 데이터의 기본적인 formation이다. 이전의 records를 이용하여 다음 target을 예측하고, step만큼 window를 sliding하면서 다음 target을 예측하는 방식을 반복한다.

**split_sequence**를 이용하여 순환 데이터를 구성하겠다. split_sequence 중에서도 **특정 window에서 마지막 레코드의 target 값을 선택** 하는 방식을 택해보자. 기존 함수는 아래와 같다.

```python
    def split_sequences(sequences, n_steps) :
        X, y = list(), list()
        for i in range(len(sequences)):
        # find the end of this pattern
            end_ix = i + n_steps
            # check if we are beyond the dataset
            if end_ix > len(sequences):
                break
            # gather input and output parts of the pattern
            seq_x, seq_y = sequences[i:end_ix, :-1], sequences[end_ix-1, -1]
            X.append(seq_x)
            y.append(seq_y)
        return np.array(X), np.array(y)
```

위 함수는 attribute의 첫번째 열부터 마지막-1 열까지를 input으로 split, 마지막 열을 target으로 split하고 있다. **이를 여기에 사용하면 당연히 안되는게, 우리는 input을 특정 window의 전체 attribute, target을 window 바로 다음 record의 Close 값으로 놓아야 하기 때문이다.** 따라서 위 함수를 변형하여 순환 데이터를 만든다. 

````python
# 순환 데이터를 구성하기 위한 split_sequence (마지막 record의 target 값 (class 값) 선택)

# def split_sequences(sequences, n_steps) :
# 	X, y = list(), list()
# 	for i in range(len(sequences)):
#     # find the end of this pattern
# 		end_ix = i + n_steps
# 		# check if we are beyond the dataset
# 		if end_ix > len(sequences):
# 			break
# 		# gather input and output parts of the pattern
# 		seq_x, seq_y = sequences[i:end_ix, :-1], sequences[end_ix-1, -1]
# 		X.append(seq_x)
# 		y.append(seq_y)
# 	return np.array(X), np.array(y)
````

````python
# 변형 버전

def split_sequences_custom(sequences, n_steps):
    X, y = list(), list()
    for i in range(len(sequences)):
        end_ix = i + n_steps

        if end_ix >= len(sequences) : 
            break

        seq_x, seq_y = sequences[i:end_ix, :], sequences[end_ix, -1] 
        '''
            seq_x = window 만큼의 모든 attribute, 
            seq_y = window 바로 다음 record의 마지막 열
        '''
        X.append(seq_x)
        y.append(seq_y)
        
    return np.array(X), np.array(y)
````

````python
# 순환 데이터 구성

'''
    step 수 : 1
    window 수 : 20
'''

X_train_seq, y_train_seq = split_sequences_custom(df_train_scaled, n_steps=20) # window 사이즈를 20으로 구성
X_test_seq, y_test_seq = split_sequences_custom(df_test_scaled, n_steps=20) # window 사이즈를 20으로 구성

print("X_train_seq:", X_train_seq.shape, "y_train_seq:", y_train_seq.shape)
print("X_test_seq:", X_test_seq.shape, "y_test_seq:", y_test_seq.shape)
````

````python
# 실제로 어떻게 구성되었는지 앞쪽 5개의 component만 출력

for idx in range (5) :
    print (f"X_train_seq[{idx}] : ", X_train_seq[idx])
    print (f"y_train_seq[{idx}] : ", y_train_seq[idx])
    
    if idx == 0 : 
        prev = X_train_seq[idx]
        continue

    if (prev[1:] == X_train_seq[idx][:-1]).all() : 
        '''
            step 1씩 이동했으니깐, 이전 X_train_seq 원소의 2번째 record ~ 마지막번째 record와
            현재 X_train_seq 원소의 첫번째 record ~ 마지막-1번째 record는 같을 수밖에 없습니다.
        '''
         
        print ("window가 step 1씩 이동하면서 순환 데이터를 만듦")
    
    prev = X_train_seq[idx]
    print()
````

### 2. 1D-CNN (time-series) using split_sequence
- **reference : https://machinelearningmastery.com/how-to-develop-convolutional-neural-network-models-for-time-series-forecasting/**



- 1D-CNN을 사용하여 다음 Close 값을 예측한다.


````python
# modeling

from tensorflow.keras.models import Model
from tensorflow.keras import layers
import keras

input_layer = keras.Input(shape=(X_train_seq.shape[1], X_train_seq.shape[2]))

# feature extractor
x = layers.Conv1D(filters=32, kernel_size=4, activation='relu', padding='same')(input_layer)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.5)(x)

x = layers.Conv1D(filters=64, kernel_size=4, activation='relu', padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.4)(x)

x = layers.Conv1D(filters=32, kernel_size=4, activation='relu', padding='same')(x)
x = layers.BatchNormalization()(x)
x = layers.Dropout(0.3)(x)

# flatten
x = layers.Flatten()(x)

# dense layer
x = layers.Dense(32, activation='relu')(x)
x = layers.Dropout(0.1)(x)
x = layers.Dense(16, activation='relu')(x)
x = layers.Dropout(0.1)(x)
x = layers.Dense(8, activation='relu')(x)

output_layer = layers.Dense(1)(x)

# Create the model
model = Model(inputs=input_layer, outputs=output_layer)

# Model summary
model.summary()

# Plot the model architecture
keras.utils.plot_model(model, show_shapes=True, show_layer_activations=True,
                       show_dtype=True, show_layer_names=True, show_trainable=True)
````

````python
# compile and fit, visualization loss

print("unique labels in y_train_seq:", np.unique(y_train_seq))

model.compile (metrics=['mae'], loss='mse', optimizer='adam')
history = model.fit (X_train_seq, y_train_seq, epochs=50, 
                     verbose=True, validation_split=0.1)

print (history)
plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title("visualization loss and validation loss")
plt.legend()
plt.show()
````

````python
# model prediction이 잘 되는지 테스트 : visualization

y_pred = model.predict(X_test_seq)

plt.figure(figsize=(14, 5))
plt.plot(test_dates[:y_test_seq.shape[0]], y_test_seq, color='green', 
         label='original price')
plt.plot(test_dates[:y_test_seq.shape[0]], y_pred, color='blue', 
         label='prediction price')
plt.title(f'1D-CNN')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
````

### 3. stacked LSTM

<img src="https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FbnMYOU%2FbtsgwMLw4rs%2F2xIg0bWNlpZqAcEUKdKpPk%2Fimg.png" width="500px">

- LSTM을 사용하여 modeling한다. LSTM layer를 여러 개 겹쳐서 구성한다.
    - 위 figure와 같이 **stacked LSTM**을 통해 prediction할 것이다.
        - 1층에서 모든 sequence의 hidden state를 출력해야 하기 때문에, hyperparameter를 return_sequences=True로 두어야 할 것이다.
        - 2층에서 마지막 sequence의 출력이 필요하기 때문에 False로 둔다.
- LSTM 이후에 Dense layer을 통한 regression formation을 만든다.

**stacked LSTM과 단일 LSTM 모델을 비교하여 누가 더 추세를 잘 prediction하는지 시각화해보자.**

````python
# modelling (stacked LSTM)

from tensorflow.keras import layers

model1 = keras.models.Sequential()
model1.add(layers.LSTM(64, input_shape=(X_train_seq.shape[1], X_train_seq.shape[2]), 
                       return_sequences=True))
model1.add(layers.LSTM(32, return_sequences=False))
model1.add(layers.Dense(1))

model1.summary()
keras.utils.plot_model(model1, show_shapes=True, show_layer_activations=True, 
                       show_dtype=True, show_layer_names=True, show_trainable=True)
````

````python
# modelling (unary LSTM)

from tensorflow.keras import layers, Model

input_layer = layers.Input(shape=(X_train_seq.shape[1], X_train_seq.shape[2]))

x = layers.LSTM(64, return_sequences=False, activation='tanh')(input_layer)
output_layer = layers.Dense(1)(x)

model2 = Model(inputs=input_layer, outputs=output_layer)

model2.summary()
keras.utils.plot_model(model2, show_shapes=True, show_layer_activations=True, 
                       show_dtype=True, show_layer_names=True, show_trainable=True)
````

````python
# compile and fit, visualization loss

print("unique labels in y_train_seq:", np.unique(y_train_seq))

model1.compile (metrics=['mae'], loss='mse', optimizer='adam')
model2.compile (metrics=['mae'], loss='mse', optimizer='adam')

history1 = model1.fit (X_train_seq, y_train_seq, epochs=50, 
                     verbose=True, validation_split=0.2)
history2 = model1.fit (X_train_seq, y_train_seq, epochs=50, 
                     verbose=True, validation_split=0.2)

print (history1, history2)

plt.figure(figsize=(10, 4))
plt.plot(history1.history['loss'], label='loss')
plt.plot(history1.history['val_loss'], label='val_loss')
plt.title("stacked LSTM : visualization loss and validation loss")
plt.legend()
plt.show()

plt.figure(figsize=(10, 4))
plt.plot(history2.history['loss'], label='loss')
plt.plot(history2.history['val_loss'], label='val_loss')
plt.title("unary LSTM : visualization loss and validation loss")
plt.legend()
plt.show()
````

````python
# model prediction이 잘 되는지 테스트 : visualization

y_pred = model1.predict(X_test_seq)

plt.figure(figsize=(14, 5))
plt.plot(test_dates[:y_test_seq.shape[0]], y_test_seq, color='green', 
         label='original price')
plt.plot(test_dates[:y_test_seq.shape[0]], y_pred, color='blue', 
         label='prediction price')
plt.title(f'stacked-LSTM')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()


y_pred = model2.predict(X_test_seq)

plt.figure(figsize=(14, 5))
plt.plot(test_dates[:y_test_seq.shape[0]], y_test_seq, color='green', 
         label='original price')
plt.plot(test_dates[:y_test_seq.shape[0]], y_pred, color='blue', 
         label='prediction price')
plt.title(f'unary-LSTM')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
````

CNN과 LSTM을 결합하여 사용한다. CNN으로 먼저 feature extraction을 진행한 후, LSTM으로 장기 기억 track을 하나 더 두어 순환형 데이터에 맞는 학습을 진행할 것이다. 이렇게 학습 시 이점은 아래와 같다고 생각된다.

- 맨 처음에 복잡도를 줄이면 찾아오는 이점
    - 먼저 CNN으로 입력 복잡성을 줄이면 LSTM이 고차원 데이터에 압도되지 않고 순차적 패턴 학습에 집중 가능하다
    - 장기적인 맥락을 유지하고 정보를 오랫동안 기억할 수 있을 것이다.
- CNN-LSTM architecture을 이용하면 영상 분석 및 시계열 분류와 같은 공간적 및 시간적 동력이 주가 되는 training에서 이점을 가져올 것이다. **(예시 : https://github.com/lky473736/learning-AI/blob/main/report/EMG/EDA_CNN-LSTM_classification_EMG.ipynb)**

````python
# modeling

from tensorflow.keras import models, layers

def create_model(input_shape, num_classes) :
    model = models.Sequential()

    # Spatial Feature Extractor
    model.add(layers.Conv1D(filters=16, kernel_size=3, 
                            activation='relu', input_shape=input_shape))
    model.add(layers.Dropout(0.1))
    
    model.add(layers.Conv1D(filters=32, 
                            kernel_size=4, activation='relu'))
    model.add(layers.Dropout(0.1))

    # Apply TimeDistributed on the LSTM layers
    model.add(layers.TimeDistributed(layers.Flatten()))  

    # Temporal Feature Extractor (LSTM)
    model.add(layers.LSTM(32, return_sequences=False))
    model.add(layers.Dropout(0.4))

    # Fully Connected Layer
    model.add(layers.Dense(8, activation='relu'))
    model.add(layers.Dense(1))

    model.compile(optimizer='adam', loss='MSE', metrics=['MAE'])

    return model

model = create_model(input_shape=(X_train_seq.shape[1], X_train_seq.shape[2]), num_classes=1)

model.summary()
tf.keras.utils.plot_model(model,
                           show_dtype=True,
                           show_layer_activations=True,
                           show_layer_names=True,
                           show_shapes=True,
                           show_trainable=True)
````

````python
# compile and fit, visualization loss

print("unique labels in y_train_seq:", np.unique(y_train_seq))

model.compile (metrics=['mae'], loss='mse', optimizer='adam')
history = model.fit (X_train_seq, y_train_seq, epochs=50, 
                     verbose=True, validation_split=0.1)

print (history)
plt.figure(figsize=(10, 4))
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.title("visualization loss and validation loss")
plt.legend()
plt.show()
````
