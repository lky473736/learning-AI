```python
import numpy as np
import pandas as pd
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
import os
```

```python
'''
    Name		Data Type	Meas.	Description
	----		---------	-----	-----------
	Sex		nominal			M, F, and I (infant)
	Length		continuous	mm	Longest shell measurement
	Diameter	continuous	mm	perpendicular to length
	Height		continuous	mm	with meat in shell
	Whole weight	continuous	grams	whole abalone
	Shucked weight	continuous	grams	weight of meat
	Viscera weight	continuous	grams	gut weight (after bleeding)
	Shell weight	continuous	grams	after being dried
	Rings		integer			+1.5 gives the age in years
'''

df = pd.read_csv("../data/abalone/abalone.csv",
                names=['Sex', 'Length', 'Diameter', 
                      'Height', 'Whole weight', 
                      'Shucked weight', 
                      'Viscera weight', 
                      'Shell weight', 
                      'Rings'])
df
```

```python
df.info()
df.describe().T
```

```python
plt.figure(figsize=(15, 4))
sns.countplot(data=df,
             x='Rings')
plt.show()
```

```python
print (set(df.isnull().sum()))
df = df.fillna(df.mean)
print (set(df.isnull().sum()))
```

```python
# Sex가 Object라 각각을 숫자로 encoding

print (df['Sex'].value_counts())

df.loc[df['Sex'] == 'M', 'Sex'] = 0
df.loc[df['Sex'] == 'I', 'Sex'] = 1
df.loc[df['Sex'] == 'F', 'Sex'] = 2

print (df['Sex'].value_counts())
```

```python
plt.figure(figsize=(9, 9))

df_corr = df.corr()
sns.heatmap(df_corr,
           annot=True,
           cmap='Blues')
```

```python
plt.figure(figsize=(8, 6))

'''
    corr (상관계수) 높은 feature끼리 시각화해봄
    Diameter - Length
    Viscera weight - Whole weight
'''

plt.subplot(211)
plt.scatter (df['Diameter'],
            df['Length'])

plt.subplot(212)
plt.scatter (df['Viscera weight'],
            df['Whole weight'])
```

```python
data = df.drop('Rings', axis=1)
target = df['Rings']

data.shape, target.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target,
                                                   test_size=0.2)

X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,
                                                 test_size=0.1)

print (X_train.shape, X_test.shape, X_val.shape)
```

```python
sns.histplot(x=X_train.iloc[:, 5], 
             color='r',
             bins=10,
             kde=True,
             element='step',
             stat='percent', 
             cumulative=False)
```

```python
from sklearn.preprocessing import StandardScaler

SS = StandardScaler()

X_train_scaled = SS.fit_transform (X_train)
X_test_scaled = SS.transform(X_test)
X_val_scaled = SS.transform(X_val)
```

```python
sns.histplot(x=X_train_scaled[:, 5],  
             color='b',
             bins=10,
             kde=True,
             element='step',
             stat='percent', 
             cumulative=False)
```

---

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, BatchNormalization
import tensorflow as tf

model = Sequential([
    Dense(256, input_shape=(8,)),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.05), 
    
    Dense(128),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.05), 
    
    Dense(64),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.05), 
    
    Dense(32),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.05), 
    
    Dense(16),
    BatchNormalization(),
    Activation('relu'),
    Dropout(0.05), 
    
    Dense(8),
    Dense(1)
])

model.summary()

tf.keras.utils.plot_model(model, 
                          show_dtype=True,
                          show_layer_activations=True, 
                          show_layer_names=True, 
                          show_shapes=True, 
                          show_trainable=True)
```

```python
model.compile(loss='mean_squared_error',
             optimizer=tf.optimizers.Adam(learning_rate=0.001), 
            metrics=['MSE'])

history = model.fit (X_train_scaled, y_train,
                    validation_data=(X_val_scaled, y_val),
                    verbose=2,
                    epochs=100)
```

```python
plt.figure(figsize=(10, 6))

plt.plot(history.history['loss'],
        c='b', label='loss')
plt.plot(history.history['val_loss'],
        c='r', label='val_loss')
plt.legend()
```

```python
model.evaluate (X_train_scaled, y_train)
model.evaluate (X_val_scaled, y_val)
model.evaluate (X_test_scaled, y_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, root_mean_squared_error, mean_absolute_error

y_pred = model.predict(X_test_scaled)

print (r2_score(y_test, y_pred))
print (mean_squared_error(y_test, y_pred))
print (root_mean_squared_error(y_test, y_pred))
print (mean_absolute_error(y_test, y_pred))
```
