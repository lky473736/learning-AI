## learning-AI101 : topic2_2

topic2_1에서는 **keras를 이용하여 모델**을 생성하였다.  
만든 모델을 **model.fit을 하기 위해선 X_train, y_train, epochs값이 필요**하다.  
**CSV 데이터를 전처리하기 위해선 pandas가 필요하다.**

```python
import tensorflow as tf
import pandas as pd
```

```python
# 학습 모델 만든 것

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation="relu"), 
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])

# model.fit(x데이터, y데이터, epochs=에폭값)
```

--------

model.fit에 들어갈 데이터는 아래와 같다.
- X_train : 정답 예측에 필요한 input data
- y_train : X_train의 정답 (label)
- epochs : 학습 반복 수

X_train는 보통 **2차원 리스트 데이터 혹은 metrics** 형태이다.

```python
dataframe = pd.read_csv("./data/gpascore.csv")
dataframe
```

X_train은 gre, gpa, rank이며 y_train은 admit일 것이다.  
X_train의 형식은 **[[380.0, 3.21, 3], [660.0, 3.67, 3] ...]이다.**  
y_train의 형식은 **[0, 1, 1, 1, 0...]이다.**

```python
# pandas를 이용하여 NULL값을 빼주는 전처리를 해보자
print(dataframe.isnull().sum()) # NULL값 갯수 세주기
# gre라는 column에 NULL값이 하나 있다
```

```python
# NULL값이 있는 row를 없애는 함수 (전처리 완료)
dataframe = dataframe.dropna()
dataframe.shape
# 원래 426개인 row가 425개가 되었다
# # data.fillna(~~) : ~~를 NULL값이 있는 모든 곳에 채워줌
```

```python
# pandas 유용한 기능
# 원하는 column만 출력
dataframe['admit']
```

```python
# 특정 column.min() or column.max() : 그 열의 최솟값 or 최댓값 리턴
# 특정 column.mean : 그 열의 평균 리턴
# 특정 column.count() : 그 열의 원소 갯수
print (dataframe['gpa'].min())
print (dataframe['gpa'].max())
print (dataframe['gpa'].mean())
print (dataframe['gpa'].count())
```

```python
# X_train, y_train 만들기
y_train = dataframe['admit'].values # .values : column 값들을 1차원 리스트로 담아줌

X_train = []
for i, rows in dataframe.iterrows() : 
    # iterrow() : df의 한 행씩 출력
    # i : index, rows : 한 행
    X_train.append ([rows['gre'], rows['gpa'], rows['rank']])

print (X_train)
```
