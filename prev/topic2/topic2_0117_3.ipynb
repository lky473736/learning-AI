```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
# 여태까지 했던 것들 
# model layer 만들기 -> model compile -> train 데이터 만들기

# 전처리 & train 데이터 만든 것

dataframe = pd.read_csv("./data/gpascore.csv")
dataframe = dataframe.dropna() # preprocessing

y_train = dataframe['admit'].values 

X_train = []
for i, rows in dataframe.iterrows() : 
    X_train.append ([rows['gre'], rows['gpa'], rows['rank']])

#######################

# 학습 모델 만든 것

model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(64, activation="relu"), 
    tf.keras.layers.Dense(128, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")])

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=['accuracy'])

# model.fit(x데이터, y데이터, epochs=에폭값)
```

-----------

**X_train, y_train을 하였으니 이제 fitting을 해야한다 (학습)**  
하지만 fit에 집어넣을 땐 **파이썬 리스트를 그냥 집어넣으면 에러가 난다**  
**numpy array 혹은 tensor**를 집어넣어야 한다

```python
# 리스트를 numpy array로 변환하기
# np.array(리스트)

model.fit (np.array(X_train), np.array(y_train), epochs=1000) # w값 찾기 == 학습
# (이차전지와인공지능) 아래를 좀 보니깐 아마도 minibatch 방식을 사용한듯?
```

--------

```python
# 모델로 예측하기 (신규데이터)
# model.predict (신규데이터값 (X_train과 같은 형태의))
# predict도 안에 np.array를 넣을 수 있다

print(model.predict (np.array([[350, 2, 3]])))
print(model.predict (np.array([[780, 4, 1]])))
```
