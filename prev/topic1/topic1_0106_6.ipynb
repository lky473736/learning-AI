총손실(loss)이 가장 적은 저 극소값을 찾기 위해 아래와 같은 식으로 점을 이동시킴  
1) 처음엔 w값 점 랜덤으로 세팅
2) w값 바탕으로 총손실(loss) 계산
3) 경사하강으로 새로운 w값 업데이트
4) 2~3 반복 (loss가 더 이상 안날때까지)


> 새로운 $w$ = 원래 $w$ - 접선의 기울기 ($\times$ learning rate (superparameter))  
> learning rate를 곱하는 이유 : overshooting 막기 위해 & local minima (극소가 여러개일 수 있음)

learning rate optimizer : 학습 중간중간에 learning rate를 수정케 하는 도구
- momentum
- adagrad
- RMSProp
- **adam** (얘를 가장 많이 씀)
- ...
