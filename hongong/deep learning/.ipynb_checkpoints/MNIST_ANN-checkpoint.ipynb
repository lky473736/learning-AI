```python
# MNIST  

from keras.datasets.fashion_mnist import load_data

(X_train, y_train), (X_test, y_test) = load_data()
print(X_train.shape, X_test.shape)
```

```python
# train set, test set scaling,  reshape 

X_train_scaled = X_train / 255.0 # 0~1     (numpy  )
X_train_scaled = X_train.reshape(-1, 28*28)

X_test_scaled = X_test / 255.0 # 0~1     (numpy  )
X_test_scaled = X_test.reshape(-1, 28*28)
```

```python
#  SGDClassifier   

from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import cross_validate
import numpy as np

sc = SGDClassifier(loss='log_loss', max_iter=20) # SGDClassifier  ( )
scores = cross_validate(sc, X_train_scaled, y_train, n_jobs=-1)

print (scores)
print ('score : ', np.mean(scores['test_score']))
```

```python
#    
#  X_train_scaled  X_train_scaled X_val_scaled ,  val evaluate  

from sklearn.model_selection import train_test_split

X_train_scaled, X_val_scaled, y_train, y_val = train_test_split(X_train_scaled, y_train,
                                                               test_size=0.2)

print (X_train_scaled.shape, X_val_scaled.shape)
```

```python
# modelling

from tensorflow import keras

dense = keras.layers.Dense(10, activation='softmax', input_shape=(784,))
model = keras.Sequential(dense)

# 10 :   (1 regression)
# softmax : 10   softmax,   sigmoid
#  28 * 28,  784 feature input layer node 
```

```python
# compile

model.compile (loss='sparse_categorical_crossentropy', metrics='accuracy')

#  target one-hot encoding     sparse loss 
# metrics accuracy  epoch  accuracy     (loss   )
```

```python
# fit

model.fit(X_train_scaled, y_train, epochs=50)
```

```python
# evaluate

model.evaluate(X_val_scaled, y_val) # validation

#   loss,   accuracy
```

```python
# predict

predict = model.predict(X_test_scaled)

cnt = 0

for i in range (len(predict)) : 
    for j in range (10) :
        if predict[i][j] != 0 : 
            print (f"prediction : {j} | real : {y_test[i]}")
            
            if j == y_test[i] :
                cnt += 1
                
print ("cnt : ", cnt)
print ("test accuracy : ", cnt/len(predict))
```
