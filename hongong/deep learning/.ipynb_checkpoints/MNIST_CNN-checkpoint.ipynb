### CNN


-              
- layer 
    - convolution layer (feature extraction) :   pooling      feature map 
    - flatten : convolution layer   feature map (pooling )   (1 )
    - dense layer : classification  dense layer 
-  
    - (1)    (  )  convolution, feature map ,  concatenate feature map 
        - convolution  padding       feature map   ,   
            -    padding='same'
            -  ( )  padding='valid'
        - stride     convolution  , n   
    - (2)  feature map pooling   
        - pooling  feature map    (  )
        - pooling 
            - max pooling : pooling   
            - average pooling : pooling     
        - pooling convolution    , stride padding  

```python
# MNIST    scaling, reshape
#     3, 3 reshape 

from tensorflow import keras
from sklearn.model_selection import train_test_split

(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()

X_train_scaled = X_train.reshape(-1, 28, 28, 1) / 255.0
X_test_scaled = X_test.reshape(-1, 28, 28, 1) / 255.0

X_train_scaled, X_val, y_train, y_val = train_test_split(X_train_scaled, y_train,
                                                       test_size=0.2)
```

```python
# modelling

model = keras.Sequential([
    keras.layers.Conv2D (32, kernel_size=(3,3), activation='relu', padding='same', input_shape=(28, 28, 1)), 
    # 32 , (3, 3)  , same padding  feature map  
    keras.layers.MaxPooling2D((2,2)),
    # (2,2)   pooling
    keras.layers.Conv2D (64, kernel_size=(3,3), activation='relu', padding='same',), 
    keras.layers.MaxPooling2D((2,2)),
    
    ################## dense layer ##################
    
    keras.layers.Flatten(),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dropout(0.4), # overfitting   dropout
    keras.layers.Dense(10, activation='softmax')
])

model.summary()
keras.utils.plot_model(model, show_shapes=True)
```

```python
# compile, fit
# overfitting   early stopping callback 

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
             metrics='accuracy') # one-hot encoding   sparse
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)

history = model.fit(X_train_scaled, y_train, epochs=50, validation_data=(X_val, y_val), 
                    callbacks=[early_stopping_cb])

history
```

```python
# validation loss train loss 

import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train loss', 'validation loss'])
plt.show()
```

```python
# validation evalutaion test evaluation

score_val = model.evaluate(X_val, y_val)
score_test = model.evaluate(X_test_scaled, y_test)

print (f"validation evaluation : {score_val}")
print (f"test evaluation : {score_test}")
```
