### CNN


- 입력 데이터를 특정 크기의 커널을 통하여 합성곱하여 유용한 특징만 도드라지게 하여 학습하는 방식 
- layer 구성
    - convolution layer (feature extraction) : 합성곱 레이어와 pooling 레이어의 교차로 중요한 특징값들만 모은 feature map을 만듦
    - flatten : convolution layer에서 최종적으로 만들어진 feature map (pooling 되어진)을 일렬로 펼침 (1차원 변환)
    - dense layer : classification을 위한 dense layer가 잇따름
- 작동 원리
    - (1) 각자 다른 필터 (가중치도 각자 다름)를 이용하여 convolution, feature map을 만들고, 이를 concatenate하여 feature map을 만듦
        - convolution할 때 padding을 어떻게 하는 지에 따라서 입력의 크기와 feature map의 크기가 같을 수도, 다를 수도 있음
            - 같게끔 하고 싶으면 padding='same'
            - 다르게 (크기가 줄어듬) 하려면 padding='valid'
        - stride 속성을 이용하여 한칸씩 이동하여 convolution하는 것이 아닌, n칸씩 이동할 수 있음
    - (2) 만들어진 feature map을 pooling하여 중요한 특성만 뽑아냄
        - pooling은 절대 feature map의 깊이를 건들이지 않음 (갯수를 건들이지 않음)
        - pooling의 종류
            - max pooling : pooling 크기에서 최댓값을 뽑아냄
            - average pooling : pooling 크기에서 각 값의 평균값을 뽑아냄
        - pooling은 convolution과 다르게 겹치는 부분이 없음, stride과 padding 조정 가능

```python
# MNIST 데이터 불러오기 및 scaling, reshape
# 이미지는 색깔 특성을 위해 3차원이므로, 3차원으로 reshape해주어야 함

from tensorflow import keras
from sklearn.model_selection import train_test_split

(X_train, y_train), (X_test, y_test) = keras.datasets.fashion_mnist.load_data()

X_train_scaled = X_train.reshape(-1, 28, 28, 1) / 255.0
X_test_scaled = X_test.reshape(-1, 28, 28, 1) / 255.0

X_train_scaled, X_val, y_train, y_val = train_test_split(X_train_scaled, y_train,
                                                       test_size=0.2)
```

```python
# modelling

model = keras.Sequential([
    keras.layers.Conv2D (32, kernel_size=(3,3), activation='relu', padding='same', input_shape=(28, 28, 1)), 
    # 32개의 필터, (3, 3)으로 되어있는 커널, same padding으로 입력과 feature map의 사이즈가 같게끔
    keras.layers.MaxPooling2D((2,2)),
    # (2,2) 크기의 최대 pooling
    keras.layers.Conv2D (64, kernel_size=(3,3), activation='relu', padding='same',), 
    keras.layers.MaxPooling2D((2,2)),
    
    ################## dense layer ##################
    
    keras.layers.Flatten(),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dropout(0.4), # overfitting을 방지하기 위한 dropout
    keras.layers.Dense(10, activation='softmax')
])

model.summary()
keras.utils.plot_model(model, show_shapes=True)
```

```python
# compile, fit
# overfitting을 막기 위한 early stopping callback 도입

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy',
             metrics='accuracy') # one-hot encoding을 안했기 때문에 sparse로
early_stopping_cb = keras.callbacks.EarlyStopping(patience=3, restore_best_weights=True)

history = model.fit(X_train_scaled, y_train, epochs=50, validation_data=(X_val, y_val), 
                    callbacks=[early_stopping_cb])

history
```

```python
# validation loss와 train loss의 시각화

import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['train loss', 'validation loss'])
plt.show()
```

```python
# validation evalutaion과 test evaluation

score_val = model.evaluate(X_val, y_val)
score_test = model.evaluate(X_test_scaled, y_test)

print (f"validation evaluation : {score_val}")
print (f"test evaluation : {score_test}")
```
