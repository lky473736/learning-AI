### decision tree
- 의사 결정 트리 : 질문 노드를 이등분의 효과가 가장 높게 배치하여 분류
- 질문 노드 배치 시 부모 노드와 자식 노드의 불순도 차이가 가장 크도록 배치
- gini impurity 혹은 엔트로피를 점수로 삼아서 배치
- 각 문제 별로 decision tree의 예측 방식
    - classification : leaf node들의 class를 조사 후, 가장 많은 class가 예측 클래스 
    - regression : leaf node들의 target을 평균한 것이 예측값
    

```python
# wine dataset을 불러옴

import pandas as pd
wine = pd.read_csv('https://bit.ly/wine_csv_data')
wine.head(), wine.info(), wine.describe()  

# 아래에서 class는 target으로, alcohol과 sugar, pH는 feature로 사용하면 될 것 같음
```

```python
# wine_input과 wine_target으로 변수 나누기

import numpy as np

wine_input = wine[['alcohol', 'sugar', 'pH']].to_numpy()
wine_target = wine['class'].to_numpy()

print (wine_input.shape, wine_target.shape)
```

```python
# train_test_split으로 train set, test set으로 split (8:2)

from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(wine_input, wine_target, test_size=0.2)
```

```python
# decision tree에서는 standardization이 필요 없다
# fit, score 

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit (X_train, y_train)

print ('train score : ', dt.score(X_train, y_train))
print ('test score : ', dt.score(X_test, y_test))

# train score가 test score보다 높아 overfitting되었음 (이를 줄이기 위해 추후에 규제해야)
```

```python
# 시각화

import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure()
plot_tree(dt)
plt.show()
```

```python
# max_depth를 지정하면 depth=max_depth만큼 시각화 가능
# filled = True를 지정하면 양수 클래스의 비율이 높을 수록 파랗게 보이게 할 수 있음
# feature_name으로 특성의 이름 전달

plot_tree (dt, max_depth=1, filled=True, feature_names=['alcohol', 'sugar', 'pH'])
plt.show()
```

```python
# overfitting을 해결하기 위하여 max_depth를 조정해가면서 모델을 만들어보자
# 반복문을 돌려서, 1부터 100까지 max_depth 조정, train score와 test score의 차이가 가장 적은 depth 확인하기

score = []

for i in range (1, 101) :
    dt = DecisionTreeClassifier(max_depth=i)
    dt.fit(X_train, y_train)
    
    current_score = [i, dt.score(X_train, y_train), dt.score(X_test, y_test)]
    score.append (current_score) # [train data score, test data score]
    print (f'depth 0~{i} | train score : {current_score[1]}, test score : {current_score[2]}')
    
score.sort (key = lambda x : abs(x[2] - x[1]))
print (score)
print ('최적의 max_depth 5위군 : ', score[:5])
```
