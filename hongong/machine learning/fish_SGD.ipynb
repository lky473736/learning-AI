### 확률적 경사 하강법 == SGD
- 기존 모델은 fit을 하는 순간 전에 학습하였던 weight와 bias들이 전부 초기화되었음
- 점진적으로 학습하기 위하여 (매번 다른 방법론을 적용 or 실시간적인 학습) SGD 도입 
- loss function이 최소가 되는 극점을 찾아, 가장 적은 양의 손실을 낼 때까지 반복적으로 학습 (epoch만큼)
    - 너무 많은 epoch : overfitting을 야기할 수 있음
    - 너무 적은 epoch : underfitting을 야기할 수 있음
- 연속적인 loss function을 사용하기 위해 logistic regression을 손실함수로 사용 (이중 분류)
- 다중 분류에서는 cross-entropy를 손실함수로 사용 
- 회귀에서는 평균 제곱 오차 혹은 평균 절댓값 오차를 사용함

```python
# fish 데이터 준비 -> species를 target으로, 나머지는 입력으로

import pandas as pd
fish = pd.read_csv ('https://bit.ly/fish_csv_data')
print(fish.head())

fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species']
```

```python
# train set, test set으로 split 
# 표준화 전처리 (둘의 scale을 같게 하기, 경사하강법을 적용하기 전 필수 과정)

from sklearn.model_selection import train_test_split 
train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)

from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input) # 항상 train_input로 학습한 통계값으로 test set 변환
```

```python
# SGDClassifier를 도입하여 확률적 경사 하강법과 함께 분류 문제를 해결

from sklearn.linear_model import SGDClassifier # 만약 회귀라면 regressor 사용하면 됨
sc = SGDClassifier(loss='log_loss', max_iter=10, random_state=42) 
# log == 로지스틱 사용한다는 뜻, max_iter == epoch 횟수 (10 정도면 100% 과소적합될 것 같음)
sc.fit(train_scaled, train_target)

print ("train score : ", sc.score(train_scaled, train_target))
print ("test score : ", sc.score(test_scaled, test_target))

# epoch가 너무 적어서 과소적합됨 -> epoch 수 늘려보기
```

```python
# 추가 학습 

sc.partial_fit(train_scaled, train_target)

print ("train score after additional fit : ", sc.score(train_scaled, train_target))
print ("test score after additional fit : ", sc.score(test_scaled, test_target))

# epoch값을 변동해가면서 최적의 epoch를 찾는 것이 낫겠음
```

```python
# 최적의 epoch를 찾기 위해 반복문을 돌려 점수를 저장

import numpy as np

train_score = []
test_score = []
sc = SGDClassifier(loss='log_loss', random_state=42)
classes = np.unique(train_target)

for iter_num in range (1, 1001) : 
    sc.partial_fit(train_scaled, train_target, classes=classes) # 처음에 fit을 하지 않았기 때문에 classes를 넣어줌
    
    train_s = sc.score(train_scaled, train_target)
    test_s = sc.score(test_scaled, test_target)
    
    print (f'{iter_num}번째 학습 | train score : {train_s}, test score : {test_s}')
    train_score.append (train_s)
    test_score.append (test_s)
```

```python
# 시각화

import matplotlib.pyplot as plt

plt.plot (train_score, 'r')
plt.plot (test_score, 'b')
plt.xticks([i for i in range (0, 1001, 100)])
plt.xlim(0, 200)
plt.show()

# 아래를 보니 epoch는 100 정도가 딱 적당한 것 같다 (train score와 test score가 거의 일치)
```
