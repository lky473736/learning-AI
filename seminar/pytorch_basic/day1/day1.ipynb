#### Linear regression implementation using Pytorch

- 순서
    - (1) X, y 데이터 생성
    - (2) a, b 초기값 설정
    - (3) loss function 설계 (오차제곱법 MSE)
    - (4) y와 f(x)의 오차 계산
    - (5) 오차 이용하여 a, b 변화량 계산
    - (6) 변화량 토대로 얼마나 값 변경할 지 결정
    - (7) a, b 변경
    - 4-7 반복하여 오차를 최소화

```python
import torch
from torch import nn  # dense layer (FCN)
from torch import optim  # optimizer
```

```python
a, b = 0.1, 0.3  # 초기값 설정

# X, y 데이터 생성
X = torch.normal(0.0, 0.55, (5000, 1))  # <-- torch.normal() == 정규분포 상에서 데이터를 뽑겠다 (normal distribution)
y = a * X + b + torch.normal(0.0, 
                             0.03, 
                             (5000, 1))  # <-- 여기서 노이즈를 조금 추가한 것임

plt.scatter (X, y)
plt.show()
```

#### 기존 방법 (sklearn)

```python
import random

X = list()
y = list()

for i in range(500):
    X.append(i + 1)
    y.append(2 * (i + 1) + random.choice([0.5, -0.5, 1, -1, 2, -2]) *
             random.randint(-500, 500))

plt.figure(figsize=(20, 8))
plt.scatter(X, y)
plt.show()
```

```python
for i, compo in enumerate (X) :
    if i % 8 == 0 : 
        print ()
    print (f"({X[i]}, {y[i]})", end=' ')
```

```python
from sklearn.linear_model import LinearRegression

X = np.array(X)
X = X.reshape(-1, 1)

y = np.array(y)
print (X.shape, y.shape)

model = LinearRegression()
model.fit(X, y)

print ("coefficient : ", model.coef_)
print ("bias : ", model.intercept_)
```

```python
plt.figure(figsize=(20, 8))
plt.scatter (X, y)
plt.plot (X, [compo*model.coef_ + model.intercept_ for compo in X], 
         c = 'r')
plt.show()
```
