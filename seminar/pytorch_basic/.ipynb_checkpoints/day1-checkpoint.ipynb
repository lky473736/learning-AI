#### Linear regression implementation using Pytorch

- 순서
    - (1) X, y 데이터 생성
    - (2) a, b 초기값 설정
    - (3) loss function 설계 (오차제곱법 MSE)
    - (4) y와 f(x)의 오차 계산
    - (5) 오차 이용하여 a, b 변화량 계산
    - (6) 변화량 토대로 얼마나 값 변경할 지 결정
    - (7) a, b 변경
    - 4-7 반복하여 오차를 최소화

```python
import torch
from torch import nn  # dense layer (FCN)
from torch import optim  # optimizer
```

```python
a, b = 0.1, 0.3  # 초기값 설정

# X, y 데이터 생성
X = torch.normal(0.0, 
                 0.55, 
                 (5000, 1))  # <-- torch.normal() == 정규분포 상에서 데이터를 뽑겠다 (normal distribution)
y = a * X + b + torch.normal(0.0, 
                             0.03, 
                             (5000, 1))  # <-- 여기서 노이즈를 조금 추가한 것임

plt.scatter (X, y)
plt.show()
```

```python
param = torch.randn(2, requires_grad=True) # <-- 랜덤한거 아무거나 2개 만들어주기

# class Model(nn.Module):
#     def __init__(self):
#         super(Model, self).__init__()
#         self.linear = nn.Linear(1, 1)

#     def forward(self, X):
#         X = self.linear(X)
#         return X

# model = nn.Linear(1, 1) # https://pytorch.org/docs/stable/generated/torch.nn.Linear.html
# model = Model()
```

```python
print(f"Initial parameters")
print(f"a: {param[0].item():.3f}, b: {param[1].item():.3f}\n") 
```

```python
LEARNING_RATE = 0.05

criterion  = nn.MSELoss() # <--- Loss function 정의
optimizer = optim.SGD([param], lr=LEARNING_RATE) # <--- SGD라고 하는 최적화알고리즘
# optimizer = optim.SGD(model.parameters(), lr=LEARNING_RATE)
```

```python
# Training loop
print("Start training")

EPOCHS = 200
for epoch in range(EPOCHS): # <--- epoch을 반복문으로 표현
    y_pred = param[0]*X + param[1] # <----- y_pred를 만듦 (여기서 param은 계속 변동하는 값)
    # y_pred = model.forward(x)

    loss = criterion(y_pred, y) # <--- 앞에는 예측값, 뒤에는 실제 정답값 (loss 계산)

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad() # optimizer 초기화
    loss.backward() # 각각의 매개변수에 대하여 편미분을 진행 (그니깐 역전파 한다는 뜻)
    optimizer.step() # 편미분값 기준으로 학습 진행됨

    if (epoch == 0) or ((epoch+1) % 10 == 0):
        # param = list(model.parameters())        
        y_pred = param[0].detach() * X + param[1].detach()
        print(f"Epoch: {epoch+1}, Loss: {loss.data.numpy():.4f}, a: {param[0].item():.3f}, b: {param[1].item():.3f}")
        
        plt.scatter(X, y)
        plt.plot (X, [compo*param[0].item() + param[1].item() for compo in X],
                 c='r')
        plt.show()
```

```python
plt.scatter(X, y)
plt.plot (X, [compo*param[0].item() + param[1].item() for compo in X],
         c='r')
plt.show()
```

#### 기존 방법 (sklearn)

```python
import random

X = list()
y = list()

for i in range(500):
    X.append(i + 1)
    y.append(2 * (i + 1) + random.choice([0.5, -0.5, 1, -1, 2, -2]) *
             random.randint(-500, 500))

plt.figure(figsize=(20, 8))
plt.scatter(X, y)
plt.show()
```

```python
for i, compo in enumerate (X) :
    if i % 8 == 0 : 
        print ()
    print (f"({X[i]}, {y[i]})", end=' ')
```

```python
from sklearn.linear_model import LinearRegression

X = np.array(X)
X = X.reshape(-1, 1)

y = np.array(y)
print (X.shape, y.shape)

model = LinearRegression()
model.fit(X, y)

print ("coefficient : ", model.coef_)
print ("bias : ", model.intercept_)
```

```python
plt.figure(figsize=(20, 8))
plt.scatter (X, y)
plt.plot (X, [compo*model.coef_ + model.intercept_ for compo in X], 
         c = 'r')
plt.show()
```
