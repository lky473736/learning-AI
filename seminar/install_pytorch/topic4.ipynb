```python
from torch import nn 

class MLP (nn.Module) :
    def __init__(self) : 
        super().__init__()
            
        self.linear = nn.Sequential(
            nn.Linear(2, 100),
            nn.Sigmoid(),
            nn.Linear(100, 1),
            nn.Sigmoid() # 끝에 Sigmoid 배치함으로써 binary classification
        )
            
    def forward(self, x) : 
        x = self.linear(x) 
        return x
    
model = MLP()
print (model)

sample = model(torch.randn(5, 2)) # 2개의 feature을 가진 record가 5개 
print(sample)
print(sample.shape)

'''
    Pytorch policy 상으로 보통 가중치 초기값은 0근처로 되어있기때문에
    sigmoid 통과하면 대부분 sig(0)인 0.5 근처에 위치하여 있다
'''
```

```python
from torch import optim 

lr = 0.001
epoch = 100

optimizer = optim.Adam(model.parameters(), 
                      lr = lr)
criterion = nn.BCELoss() # BinaryCrossEntropy


loss_history = []

model.train() # <--- train mode로 전환시키기
for ep in range (epoch) : 
    # inference
    y_hat = model(X)

    # loss
    loss = criterion(y_hat, y)
    
    # deriviate
    optimizer.zero_grad() # gradient 누적을 막기 위한 초기화
    loss.backward()
    
    # update
    optimizer.step() # 진정한 의미의 gradient update
    
    loss_history.append (loss.item())
    
    print (f"epoch : {ep+1}, train loss : {loss.item():.3f}")

plt.plot([ep for ep in range (1, epoch+1)],
        loss_history)
plt.show()
```
