### 3-1강. nn.Linear()

```python
import torch
from torch import nn

x = torch.tensor([1.])
model = nn.Linear(1, 1) # 입력 node 1개, 출력 node 1개 FCNN
print (model)
print (model.weight) # 처음에는 당연히 랜덤값
print (model.bias) # 당연히 처음에는 랜덤값

'''
Parameter containing:
tensor([[-0.2881]], requires_grad=True)
Parameter containing:
tensor([-0.9676], requires_grad=True)

requires_grad가 true인 이유 : backward로 back-propagation 편미분 진행할 gradient이기 때문
'''

y1 = model(x) # 일종의 Linear은 선형회귀. Wx + b. model은 일종의 function
print (y1)

y2 = x @ model.weight + model.bias 
print (y2)

from torchviz import make_dot
make_dot (y1)
```

```python
# class로 모델을 많이 선언한다
# Version 1.

class model(nn.Module) : 
    def __init__(self) : 
        super().__init__()
        
        # init에서 원하는 layer 정의해놓기
        self.fc1 = nn.Linear(5, 10)
        self.fc2 = nn.Linear(10, 20)
        self.fc3 = nn.Linear(20, 10)
        self.act = nn.Sigmoid() # activation function
        
    def forward (self, x) : 
        # forward에서 network 설계
        x = self.fc1(x)
        x = self.act(x)
        x = self.fc2(x)
        x = self.act(x)
        x = self.fc3(x)
        x = self.act(x)
        return x

model = model()
x = torch.randn(10, 5)
y = model(x) 
# 위 y=model(x)는 model.forward(x)랑 똑같은 기능
print (y)

make_dot(y)
```

```python
# Version 2.


class model(nn.Module) : 
    def __init__(self) : 
        super().__init__()
        
        self.FCNN = nn.Sequential(
            nn.Linear(5, 10),
            nn.Sigmoid(),
            nn.Linear(10, 20),
            nn.Sigmoid(),
            nn.Linear(20, 10),
            nn.Sigmoid()
        )
        
    def forward (self, x) : 
        x = self.FCNN(x)
        return x

model = model()
x = torch.rand(10, 5)
y = model.forward(x) 
print (y)

make_dot(y)
```

### 3-2강. linear regression using pytorch

- linear regression
    - 입력과 출력 간 관계를 선형함수로 분석
    - 데이터의 분포를 가장 잘 표현 (distance, 즉 오차) 할 수 있는 직선 찾기
    - 데이터의 추세를 파악하는 것
    - 최적의 weight와 bias를 찾기 위해
        - (1) normal equation을 이용하여 weight, bias 찾을 수 있음
        - (2) deep learning 이용
            - gradient descent를 이용하여 loss를 최소화하는 것 (직선과 데이터포인트 사이의 distance를 최소화하는 직선)
            - MAE, MSE 둘 다 구현해보자

```python
import matplotlib.pyplot as plt

x = torch.tensor([150, 160, 170, 175, 185.]) # 키
y = torch.tensor([55, 70, 64, 80, 75.])

plt.scatter(x, y)
plt.show()
```

```python
# 선형 회귀 계산 (Normal Equation 기반, 편차 이용)

w = torch.sum((x - x.mean()) * (y - y.mean())) / torch.sum((x - x.mean())**2)
b = y.mean() - w * x.mean()

print (w, b)

dots = torch.tensor([i * w + b for i in x])

plt.scatter(x, y)
plt.plot(x, dots, color='red')
plt.show()
```

```python
# brute-force 방식

a = 0.5 + torch.linspace(-0.5, 0.5, 200) # weight
b = -30 + torch.linspace(-50, 50, 200) # bias

loss = torch.zeros(len(a), len(b)) # a by b 행렬, loss 저장 예정

for i in range (len(a)) :
    for j in range (len(b)) :
        # Loss를 계속 더함 (모든 데이터포인트에 대하여)
        # 현재 loss + 예측값과 실제값의 MSE (not n division)
        for n in range (len(x)) : # 데이터 갯수
            loss[i, j] = loss[i, j] + (y[n] - (a[i] * x[n] + b[j]))**2
loss = loss / len(x)

plt.figure(figsize=[10, 9])
ax = plt.axes(projection="3d")
A, B = torch.meshgrid(a, b)
ax.plot_surface(A, B, loss)
ax.set_xlabel('a'); ax.set_ylabel('b')
ax.set_zlim([0, 1000])

plt.figure()
plt.contour(a, b, loss, 30)
plt.xlabel('a'); plt.ylabel('b'); plt.grid()
```

```python
print (loss)
print (loss.shape)
print (torch.min(loss))
w_opt = A[loss==loss.min()]
b_opt = B[loss==loss.min()]
print (w_opt, b_opt)
```

```python
# gradient descent를 이용한 gradient 찾기

# 초기값 설정
a = torch.tensor([0.5], 
                requires_grad=True) # 기울기 초기값
b = torch.tensor([-30.], 
                requires_grad=True) # 절편 초기값

learning_rate = 0.00001
epoch = 100

loss_history = []

for ep in range (epoch) : 
    # (1) prediction, inference (먼저 예측값을 만든다)
    y_hat = a * x + b 
    
    loss = 0
    # (2) loss를 구한다 (MSE)
    for i in range (len(x)) : 
        loss += (y[i] - y_hat[i])**2
    loss = loss / len(x)
    loss_history.append (loss.item())
    
    # (3) loss를 이용하여 backpropagation
    loss.backward() # autograd로 모든 leaf tensor에 대하여 편미분값 구해짐
    
    # (4) weight를 update한다
    with torch.no_grad() : 
        a -= learning_rate * a.grad # 경사하강법 공식
        b -= learning_rate * b.grad 
#     a.grad = torch.tensor([0.])<-- gradient 초기화 이런식으로하면 에러남
#     b.grad = torch.tensor([0.])
    if a.grad is not None:
        a.grad.zero_()
    if b.grad is not None:
        b.grad.zero_()

    
    print (f'epoch {ep+1} : train loss : {loss}')
    print (f'current weight : {a} and bias : {b}')
    print (' - ' * 15)

plt.plot ([ep+1 for ep in range(epoch)],
         loss_history)
plt.show()
```
